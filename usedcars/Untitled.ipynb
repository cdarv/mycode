```python
import pandas as pd
```

```python
cars1 = pd.read_csv('cars1.csv')
cars2 = pd.read_csv('cars2.csv')
```

```python
df = pd.concat([cars1,cars2])
df.set_index('Stock #', inplace = True)
df = df[df['Photo Count'] != 0]
df
```

```python
df.to_excel("AllCars.xlsx", sheet_name='Inventory')  
```

```python
topdf = df.head(10)
topdf.sort_values(['Price'], ascending = False)
```
