```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('~/mycode/Starbucks/starbucks.csv')
df.rename(columns={ df.columns[0]: "Item" }, inplace = True)
df = df[df["Calories"].str.contains("-")==False]
df['Calories'] = pd.to_numeric(df['Calories'])
```

```python
pd.set_option('display.max_rows', None)
df = df.sort_values(['Calories'], ascending = False).head(10)
```

```python
df
```

```python
df = df[["Item","Calories"]]
```

```python
df
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Drinks = df["Item"]
Calories = df["Calories"]
ax.bar(Drinks,Calories)
plt.xticks(rotation = 90)

plt.show()
```
